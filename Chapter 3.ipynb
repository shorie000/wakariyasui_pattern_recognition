```python
#図3.2 観測結果と事後確率の関係（n=10の時）
import matplotlib.pyplot as plt
n = 10#Number of trials

pi = np.array([0.1, 0.4, 0.5])#Content rate
theta = np.array([0.8, 0.6, 0.3])#Probability of  front is appearing

r=np.arange(0,n)#Number of  front is appearing

p_all=0#probability
for i in range(3):
    p_all += pi[i]*theta[i]**r*(1-theta[i])**(n-r)

P=[]
for i in range(3):
    P.append(pi[i]*theta[i]**r*(1-theta[i])**(n-r)/p_all)

for tmp in range(len(P)):
    plt.plot(P[tmp], label=str(tmp))
    plt.legend()
    plt.ylabel("Posterior probability")
    plt.xlabel("Number of occurrences of front")
    plt.ylim(0, 1)
plt.show()
```

```python
#図3.4　観測回数とベイズ誤り回数(本文図と一致しないが、n=10, n=100の場合は一致)
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import comb #Combination

pi = np.array([0.1, 0.4, 0.5]) #Content rate
theta = np.array([0.8, 0.6, 0.3]) #Probability of  front is appearing

eB_all=[]
for n in range(0, 101):# n: Number of trials
    eB=0 #e_B
    for r in range(0, n+1):#Number of  front is appearing
        Px_w = theta**r*(1-theta)**(n-r)# P(x|ω)
        Pw_x = Px_w * pi / np.sum(Px_w*pi)# P(ω|x)
        eB_r = min(1-Pw_x)#e_B(r)
        Pn_r = np.sum(Px_w*pi*comb(n, r))# P_n(r)
        eB += eB_r*Pn_r # e_B
    eB_all.append(eB)
plt.plot(eB_all)
plt.xlabel("Number of trials")
plt.ylabel("Bayesian error probability")
plt.show()
```
