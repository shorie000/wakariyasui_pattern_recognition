```python
#図2.2　
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt

thetas = np.array([0.25, 0.5, 0.9])
n=10
rs = range(0, n+1)
graph = []
for theta in thetas:
    graph_tmp = []
    for r in rs:
        graph_tmp.append(comb(n, r)*(1-theta)**(n-r)*theta**r)
    graph.append(graph_tmp)
for tmp in graph:
    plt.plot(rs, tmp)
    plt.ylim(0, 0.5)
plt.show
```

```python
#図2.3
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import random

thetas = np.array([0.8, 0.6, 0.3])
thetas_T = 1-thetas
pis = np.array([0.1, 0.4, 0.5])
n=100

pre=0.8

P = np.zeros((3,n))
if random.random()<=pre:
    for i in [0,1,2]:
        P[i,0] = pis[i]*thetas[i]/np.sum((thetas*pis))
else:
    for i in [0,1,2]:
        P[i,0] = pis[i]*thetas_T[i]/np.sum((thetas_T*pis))


    
for i in range(1,n):
    if random.random()<=pre:
        for j in [0,1,2]:
            P[j,i] = thetas[j]*P[j,i-1]/np.sum((thetas*P[:,i-1]))
    else:
        for j in [0,1,2]:
            P[j,i] = thetas_T[j]*P[j,i-1]/np.sum((thetas_T*P[:,i-1]))
    

for tmp in range(P.shape[0]):
    plt.plot(P[tmp, :], label=str(tmp))
    plt.ylim(0, 1)
    plt.legend()
plt.show
```
